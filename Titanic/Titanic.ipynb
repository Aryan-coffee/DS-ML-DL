```python
# Load built-in Titanic dataset from seaborn
df=sns.load_dataset('titanic')
print('shape:',df.shape)
df.head(10)
```

```python
# Count missing values
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)

missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_pct
}).query('`Missing Count` > 0').sort_values('Missing %', ascending=False)

print(missing_df)
```

```python
# Visualize missing values as heatmap
plt.figure(figsize=(12, 4))
sns.heatmap(df.isnull(), yticklabels=False, cbar=False,
            cmap='viridis', linewidths=0)
plt.title('Missing Value Map (yellow = missing)', fontsize=13)
plt.tight_layout()
plt.show()
```

```python
#Data Cleaning 
df['age'] = df.groupby(['pclass','sex'])['age'].transform(
    lambda x: x.fillna(x.median())
)
print(f"Age nulls remaining : {df['age'].isnull().sum()}")
```

```python
 #Fill Embarked with mode
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
print(f"Embarked nulls remaining: {df['embarked'].isnull().sum()}")
```

```python
#  Drop 'deck' column (77%+ missing) and 'embark_town' (redundant)
df.drop(columns=['deck', 'embark_town', 'alive'], inplace=True, errors='ignore')
print("Dropped: deck, embark_town, alive")
```

```python
# Check duplicates
print(f"Duplicate rows: {df.duplicated().sum()}")

# Confirm no more critical nulls
print(f"\nRemaining nulls:\n{df.isnull().sum()[df.isnull().sum()>0]}")
```

```python
# Feature Engineering
df['family_size'] = df['sibsp'] + df['parch'] + 1
df['is_alone']    = (df['family_size'] == 1).astype(int)
df['age_group']   = pd.cut(df['age'],
    bins=[0, 12, 18, 35, 60, 100],
    labels=['Child', 'Teen', 'Adult', 'Middle-aged', 'Senior'])
df['fare_band']   = pd.qcut(df['fare'], 4,
    labels=['Low', 'Mid', 'High', 'Premium'])

print("✅ Feature engineering done!")
print(df[['family_size','is_alone','age_group','fare_band']].head(8))
```

```python
# Final cleaned dataset shape & dtypes
print(f"Final shape: {df.shape}")
df.dtypes
```

```python
#EDA
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Survival Overview', fontsize=15, fontweight='bold', y=1.02)

# (A) Survival count
surv_count = df['survived'].value_counts()
axes[0].bar(['Not Survived', 'Survived'], surv_count.values,
            color=['#d95f4b', '#2a7d6f'], edgecolor='white', width=0.5)
for i, v in enumerate(surv_count.values):
    axes[0].text(i, v + 5, str(v), ha='center', fontweight='bold', fontsize=12)
axes[0].set_title('Survival Count')
axes[0].set_ylabel('Passengers')

# (B) Survival by Sex
sex_surv = df.groupby('sex')['survived'].mean() * 100
axes[1].bar(sex_surv.index, sex_surv.values,
            color=['#c9943c', '#2a7d6f'], width=0.4, edgecolor='white')
for i, (idx, v) in enumerate(sex_surv.items()):
    axes[1].text(i, v + 1, f'{v:.1f}%', ha='center', fontweight='bold')
axes[1].set_title('Survival Rate by Sex')
axes[1].set_ylabel('Survival Rate (%)')
axes[1].set_ylim(0, 105)

# (C) Survival by Pclass
pclass_surv = df.groupby('pclass')['survived'].value_counts(normalize=True).unstack() * 100
pclass_surv.plot(kind='bar', stacked=True, ax=axes[2],
                 color=['#d95f4b','#2a7d6f'], edgecolor='white', width=0.5, rot=0)
axes[2].set_title('Survival % by Passenger Class')
axes[2].set_xlabel('Passenger Class')
axes[2].set_ylabel('Percentage')
axes[2].legend(['Not Survived','Survived'], loc='upper right', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# KDE by survival
for s, label, color in zip([0,1], ['Not Survived','Survived'], ['#d95f4b','#2a7d6f']):
    sns.kdeplot(df[df['survived']==s]['age'], ax=axes[0],
                fill=True, alpha=0.4, color=color, linewidth=2, label=label)
axes[0].axvline(df['age'].median(), color='navy', ls='--', lw=1.5,
                label=f"Median Age ({df['age'].median():.0f})")
axes[0].set_title('Age Distribution by Survival (KDE)', fontsize=12)
axes[0].set_xlabel('Age'); axes[0].set_ylabel('Density')
axes[0].legend()

# Histogram
axes[1].hist(df[df['survived']==0]['age'], bins=30, alpha=0.6,
             color='#d95f4b', label='Not Survived', edgecolor='white')
axes[1].hist(df[df['survived']==1]['age'], bins=30, alpha=0.6,
             color='#2a7d6f', label='Survived', edgecolor='white')
axes[1].set_title('Age Histogram by Survival', fontsize=12)
axes[1].set_xlabel('Age'); axes[1].set_ylabel('Count')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(17, 5))

# Boxplot by class
data_bp = [df[df['pclass']==c]['fare'].values for c in [1,2,3]]
bp = axes[0].boxplot(data_bp, patch_artist=True, widths=0.5,
                     medianprops=dict(color='white', linewidth=2.5))
for patch, color in zip(bp['boxes'], ['#c9943c','#2a7d6f','#d95f4b']):
    patch.set_facecolor(color); patch.set_alpha(0.85)
axes[0].set_xticklabels(['1st Class','2nd Class','3rd Class'])
axes[0].set_title('Fare Distribution by Class')
axes[0].set_ylabel('Fare (£)')

# Violin by survived
sns.violinplot(data=df[df['fare']<df['fare'].quantile(0.97)],
               x='pclass', y='fare', hue='survived',
               split=True, inner='quartile', ax=axes[1],
               palette={0:'#d95f4b', 1:'#2a7d6f'}, linewidth=1)
axes[1].set_title('Fare Violin (Pclass × Survival)')
axes[1].set_xlabel('Passenger Class'); axes[1].set_ylabel('Fare')
axes[1].legend(['Not Survived','Survived'], fontsize=8)

# Fare band survival rate
fb_surv = df.groupby('fare_band', observed=True)['survived'].mean() * 100
axes[2].bar(fb_surv.index.astype(str), fb_surv.values,
            color=['#5a6a7e','#c9943c','#2a7d6f','#1a2744'], edgecolor='white', width=0.55)
for i, v in enumerate(fb_surv.values):
    axes[2].text(i, v+1, f'{v:.0f}%', ha='center', fontweight='bold', fontsize=10)
axes[2].set_title('Survival Rate by Fare Band')
axes[2].set_ylabel('Survival Rate (%)')
axes[2].set_ylim(0, 110)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(17, 5))

# Family size survival
fs_surv = df.groupby('family_size')['survived'].mean() * 100
axes[0].bar(fs_surv.index, fs_surv.values, color='#2a7d6f', edgecolor='white', width=0.6)
axes[0].plot(fs_surv.index, fs_surv.values, color='#d95f4b',
             marker='o', linewidth=2, markersize=8, zorder=5)
axes[0].set_title('Survival Rate by Family Size')
axes[0].set_xlabel('Family Size (incl. self)')
axes[0].set_ylabel('Survival Rate (%)')

# Alone vs not alone
alone_surv = df.groupby('is_alone')['survived'].mean() * 100
axes[1].bar(['Travelling
Alone', 'With
Family'], alone_surv.values,
            color=['#d95f4b','#2a7d6f'], edgecolor='white', width=0.45)
for i, v in enumerate(alone_surv.values):
    axes[1].text(i, v+1, f'{v:.1f}%', ha='center', fontweight='bold', fontsize=12)
axes[1].set_title('Survival: Alone vs With Family')
axes[1].set_ylabel('Survival Rate (%)')
axes[1].set_ylim(0, 110)

# Embarkation pie
emb = df['embarked'].value_counts()
colors_e = ['#1a2744','#c9943c','#2a7d6f']
wedges, _, autotexts = axes[2].pie(
    emb.values, labels=['Southampton','Cherbourg','Queenstown'],
    autopct='%1.1f%%', colors=colors_e, startangle=140,
    wedgeprops=dict(edgecolor='white', linewidth=2))
for at in autotexts: at.set_fontweight('bold')
axes[2].set_title('Embarkation Port')

plt.tight_layout()
plt.show()
```
